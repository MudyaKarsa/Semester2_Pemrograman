```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def my_line(x, m, q):
    y = m * x + q
    return y


my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')

my_dataset1 = my_dataset[my_dataset.Epoch == 'three']
my_dataset2 = my_dataset[my_dataset.Epoch == 'three-b']

x = np.linspace(52.5, 62, 60)
y = my_line(x, m=0.3, q=-10.3)

fig, ax = plt.subplots()

ax.scatter(my_dataset1.SiO2, my_dataset1.K2O, marker='o', color="#cac481", edgecolor='#000000', label=r'$3^{rd}$ Epoch')
ax.scatter(my_dataset2.SiO2, my_dataset2.K2O, marker='o', color="#141ad3", edgecolor='#000000', label=r"3-B's Epoch")
ax.plot(x, y, color="#fcf92e")

ax.annotate(r'Berapa 1$\sigma$ dari titik ini?', xy=(52.3, 7.0), xytext=(47, 8.8), fontname="times new roman", arrowprops=dict(arrowstyle="->", connectionstyle="arc3",))
ax.text(52.4, 5.6, r'$ Na_2O = 0.3 \cdot SiO_2 -10.3$', dict(size=10, rotation=33))

ax.text(55.5, 5.4, r'$ \mu_{SiO_2} = \frac {a_{1}+a_{2}+\cdots +a_{n}}{n}$ = ' + '{:.1f} [wt.%]'.format(57.721),  dict(size=11.5))
ax.set_title("Saos's $4^{th}$ Graphic, Reprise", dict(fontname="times new roman", color="#BA946B"))
ax.set_xlabel(r'SiO$_2$ [wt%]')
ax.set_ylabel(r'K$_2$O [wt%]')

ax.legend()
```

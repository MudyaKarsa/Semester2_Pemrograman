```python
import matplotlib.pyplot as plt
import pandas as pd
my_dataset1 = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# Menetapkan 2 sub-dataset, Zr > 450 dan Zr < 450.
my_sub_dataset1 = my_dataset1[my_dataset1.Zr > 450] 
my_sub_dataset2 = my_dataset1[my_dataset1.Zr < 450]


fig, ax = plt.subplots() 
  
x1 = my_sub_dataset1.Zr
y1 = my_sub_dataset1.Th
ax.scatter(x1, y1, color='blue', label="Zr > 450 [ppm]") # Untuk Zr > 450 (Biru)

x2 = my_sub_dataset2.Zr
y2 = my_sub_dataset2.Th
ax.scatter(x2, y2, color='red', label="Zr < 450 [ppm]") # Untuk Zr < 450 (Merah)

ax.set_title("Saos's Diagram Numero Dos")
ax.set_xlabel("Zr [ppm]")
ax.set_ylabel("Th [ppm]")
# Menambahkan legenda (Sudut kanan atas)
ax.legend()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
my_dataset1 = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

fig, ax = plt.subplots() 

my_data1 = my_dataset1[(my_dataset1.Epoch == 'one')] # Ketentuan ditambahkan. Mengambil data Zr dan Th dari baris yang berisi Epoch One.
ax.scatter(my_data1.Zr, my_data1.Th, label='Epoch 1')

my_data2 = my_dataset1[(my_dataset1.Epoch == 'two')] # Epoch Two.
ax.scatter(my_data2.Zr, my_data2.Th, label='Epoch 2')

my_data3 = my_dataset1[(my_dataset1.Epoch == 'three')] #Epoch Three.
ax.scatter(my_data3.Zr, my_data3.Th, label='Epoch 3')

my_data4 = my_dataset1[(my_dataset1.Epoch == 'three-b')] # Epoch Three-b.
ax.scatter(my_data4.Zr, my_data4.Th, label='Epoch 3b')
    
ax.set_title("Saos's Diagram Numero Tres")   
ax.set_xlabel("Zr [ppm]")
ax.set_ylabel("Th [ppm]")
ax.legend()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
my_dataset1 = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

epochs = ['one','two','three','three-b'] # Mendeklarasikan nilai dari variabel baru, Epochs.

fig, ax = plt.subplots() 
for epoch in epochs:
    my_data = my_dataset1[(my_dataset1.Epoch == epoch)] # Menetapkan label Epoch untuk nilai variabel epoch yang dimasukkan.
    ax.scatter(my_data.Zr, my_data.Th, label="Epoch " + epoch) # Membuat grafik
                                                # Label: Epoch + nilai epoch (one, two, three, three-b)
ax.set_title("Saos's Diagram Numero Tres 2: Electric Boogaloo")
ax.set_xlabel("Zr [ppm]")
ax.set_ylabel("Th [ppm]")
ax.legend()

```

```python
import matplotlib.pyplot as plt
import numpy as np

def ec(f, d, c0):
    cl = c0/(d * (1-f) + f)
    return cl

my_f = np.linspace(0.3, 1, 8)

my_c = ec(f=my_f, d=0.1, c0=100)

fig, ax = plt.subplots() 
ax.plot(my_f, my_c, label="Eq cryst. D = 0.1")

ax.set_title("Saos's Curve Numero Uno")
ax.set_xlabel('F')
ax.set_ylabel('C [ppm]')
ax.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def ec(f, d, c0):
    cl = c0/(d * (1-f) + f)
    return cl

my_f = np.linspace(0.3,1, 8)

my_c1 = ec(f=my_f, d =0.1, c0=100)
my_c2 = ec(f=my_f, d=1, c0=100)
my_c3 = ec(f=my_f, d=2, c0=100)

fig, ax = plt.subplots() 
ax.plot(my_f, my_c1, label="Eq cryst. D = 0.1")
ax.plot(my_f, my_c2, label="Eq cryst. D = 1")
ax.plot(my_f, my_c3, label="Eq cryst. D = 2")

ax.set_title("Saos's Curve Numero Dos")
ax.set_xlabel('F')
ax.set_ylabel('C [ppm]')
ax.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def ec(f, d, c0):
    cl = c0/(d * (1-f) + f)
    return cl

my_f = np.linspace(0.3,1, 8)

d = [0.1, 1, 2]

fig, ax = plt.subplots() 

for my_d in d:
    my_c = ec(f=my_f, d=my_d, c0=100)
    ax.plot(my_f, my_c, label='Eq cryst. D = ' +  str(my_d))

ax.set_title("Saos's Curve Numero Tres")
ax.set_xlabel('F')
ax.set_ylabel('C [ppm]')
ax.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm

my_data = pd.read_csv('DEM.csv')
fig, ax = plt.subplots() 
ax.scatter(x = my_data.X_LOC.values, 
           y = my_data.Y_LOC.values, 
           c=my_data.ELEVATION.values, 
           s=2, cmap='hsv', linewidth=0, marker='o')
ax.set_title("Saos's Topographical Map")
ax.axis('equal')
ax.axis('off')
colorbar = fig.colorbar(cm.ScalarMappable(cmap='hsv'), extend='max', ax=ax)
colorbar.set_label('Ketinggian [Meter]', rotation=270, labelpad=20)
colorbar.outline.set_edgecolor('Grey')
```

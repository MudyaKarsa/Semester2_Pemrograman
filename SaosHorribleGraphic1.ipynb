```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def my_line(x, m, q):
    y = m * x + q
    return y

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')

my_dataset1 = my_dataset[my_dataset.Epoch == 'two']
my_dataset2 = my_dataset[my_dataset.Epoch == 'three']

x = np.linspace(49, 62, 800)
y = my_line(x, m=0.2, q=-9)

fig, ax = plt.subplots()

ax.scatter(my_dataset1.SiO2, my_dataset1.K2O, marker='o', color="#ec0cd2", edgecolor="#1218D7", label=r'Epoch Numero Dos')
ax.scatter(my_dataset2.SiO2, my_dataset2.K2O, marker='o', color="#05eb46", edgecolor="#F36609", label=r'Epoch Numero Tres')
ax.plot(x, y, color="#ff7b00")

ax.annotate(r'Yang ini ketinggalan?', xy=(52.4, 7.0), color="#be0707", xytext=(47, 8.8), arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
ax.text(49, 8.7, r'Featuring Dante from Devil May Cry Series!', dict(size=30, rotation=33), color="#F2FA00", fontname="Comic Sans MS")

ax.text(57.6, 2, r'$ e =' + '{:.1f} Dollar Poundsterling Rupiah'.format(57.721),  dict(size=23), rotation=170)
ax.set_title("Saos's Graphic Numero Quatro.", fontname="times new roman")
ax.set_xlabel(r'Silicon Dioxide [wt%]', fontname="times new roman")
ax.set_ylabel(r'Pendapatan Pajak Per Kapita Kabupaten Sinjai Barat')
ax.text(60, 11, r'Tuhan, tolonglah hambamu yang tertindas aksara.', dict(size=17), fontname="times new roman")

ax.legend()
```
